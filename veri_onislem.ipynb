```python
import os
import shutil
from PIL import Image

def convert_images(source_dir, target_dir):
    # Kaynak ve hedef dizinleri tanımla

    # Hedef dizini oluştur
    if not os.path.exists(target_dir):
        os.makedirs(target_dir)

    # İşlem yapılacak klasörleri tanımla
    folders = ['test', 'train']

    # Her bir klasördeki resimleri png formatına dönüştür ve hedef dizine kopyala
    for folder in folders:
        source_folder = os.path.join(source_dir, folder)
        target_folder = os.path.join(target_dir, folder)

        # Hedef klasörü oluştur
        if not os.path.exists(target_folder):
            os.makedirs(target_folder)

        # Tüm resim dosyalarını al
        subfolders = ['images', 'masks']
        for subfolder in subfolders:
            current_source_folder = os.path.join(source_folder, subfolder)
            current_target_folder = os.path.join(target_folder, subfolder)

            # Hedef alt klasörü oluştur
            if not os.path.exists(current_target_folder):
                os.makedirs(current_target_folder)

            images = [f for f in os.listdir(current_source_folder) if f.endswith('.jpg') or f.endswith('.png') or f.endswith('.bmp')]

            # Her bir resmi png formatına dönüştür ve hedef dizine kopyala
            for image_name in images:
                # Resmi aç
                image_path = os.path.join(current_source_folder, image_name)
                img = Image.open(image_path)

                # Hedef dosya adını oluştur
                target_image_path = os.path.join(current_target_folder, image_name.split('.')[0] + '.png')

                # Resmi png olarak kaydet
                img.save(target_image_path, 'PNG')
                print(f"{image_name} converted and saved as {os.path.basename(target_image_path)}")

    print("Conversion complete.")

# Kaynak ve hedef dizinlerini tanımla
source_dir = r'C:\Users\emnkr\OneDrive\Masaüstü\SON VERİSETİ\unet'
target_dir = r'C:\Users\emnkr\OneDrive\Masaüstü\SON VERİSETİ\unet1'

convert_images(source_dir, target_dir)
```

```python
from PIL import Image
import os

def resize_images(source_dir, target_size=(512, 512)):
    # Kaynak dizinini tanımla
    test_images_dir = os.path.join(source_dir, 'test', 'images')
    test_masks_dir = os.path.join(source_dir, 'test', 'masks')
    train_images_dir = os.path.join(source_dir, 'train', 'images')
    train_masks_dir = os.path.join(source_dir, 'train', 'masks')

    # Resimleri ve maskeleri yeniden boyutlandır
    resize_images_in_directory(test_images_dir, target_size)
    resize_images_in_directory(test_masks_dir, target_size)
    resize_images_in_directory(train_images_dir, target_size)
    resize_images_in_directory(train_masks_dir, target_size)

def resize_images_in_directory(directory, target_size):
    # Dizin içindeki tüm dosyaları bul
    for filename in os.listdir(directory):
        filepath = os.path.join(directory, filename)
        # Resmi aç
        img = Image.open(filepath)
        # Resmi yeniden boyutlandır
        resized_img = img.resize(target_size, Image.BILINEAR)  # Yeniden boyutlandırma için BILINEAR filtresini kullan
        # Yeniden boyutlandırılmış resmi kaydet
        resized_img.save(filepath)
        print(f"{filename} resized to {target_size}")

# Kaynak dizinini tanımla
source_dir = r'C:\Users\emnkr\OneDrive\Masaüstü\SON VERİSETİ\unet1'

# Resimleri ve maskeleri belirtilen boyuta yeniden boyutlandır
resize_images(source_dir)

```
